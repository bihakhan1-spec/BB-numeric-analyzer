## Exercise 1 — Load the data
Load `data/sports_scores.csv` using `np.genfromtxt` with `delimiter=","` and `names=True`.

```python
# YOUR CODE HERE
data = np.genfromtxt("sports_scores.csv", delimiter=",", names=True)
```

## Exercise 2 — Points summary
For the `points_scored` column, print the average, highest, and lowest points.

```python
# YOUR CODE HERE
points_scored = data['points_scored']

print("Match summary")
print("  average :", round(points_scored.mean(), 1), "points")
print("  highest :", points_scored.max(), "points")
print("  lowest :", points_scored.min(), "points")
```

## Exercise 3 — Winning games
A game is a WIN when `points_scored` is greater than `points_allowed`.
1. Build a boolean mask for wins.
2. Count how many games were won.
3. Print the match numbers that were wins.

```python
points_allowed = data['points_allowed']

print("Match summary")
print("  average :", round(points_allowed.mean(), 1), "points")
print("  highest :", points_allowed.max(), "points")
print("  lowest :", points_allowed.min(), "points")
```

```python
# How many games won? 
points_scored = (points_scored > points_allowed).sum()
print("Number of games won:", int(points_scored))
```

```python
# Which days scores were higher than opponent team?
games_won = data['match'][points_scored > points_allowed]
print("Matches where points scored were higher than the opponent:")
print(games_won.astype(int))
```

## Exercise 4 — Best game
Use `argmax` to find the position of the highest-scoring game, then print which match number it was and how many points were scored.

```python
# YOUR CODE HERE

# Which day the score was highest? (argmax gives the POSITION of the max)
highest_pos = points_scored.argmax()
print(highest_pos)
highest_score = data['match'][highest_pos]
print("Highest score was on day", int(highest_score), "at", points_scored[highest_pos])


```

## Exercise 5 — Label the games
Use `np.where` to label each game `"HIGH"` if points_scored ≥ 105, else `"LOW"`. Print the labels.

```python
# YOUR CODE HERE
# Label each day HIGH / LOW using np.where
labels = np.where(points_scored >= 105, "HIGH", "LOW")
for m, p, lab in zip(data['match'].astype(int), points_scored, labels):
    print(f"Match {m:2d}: {p:.0f} -> {lab}")
```
