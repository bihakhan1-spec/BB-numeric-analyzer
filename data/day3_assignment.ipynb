## Task 2 — Summary statistics
Pick one important numeric column (e.g., `temp_c` or `points_scored`).
Print its mean, min, max, and standard deviation with clear labels.

```python
# YOUR CODE HERE
points_allowed = data['points_allowed']

print("Match summary")
print("  average :", round(points_allowed.mean(), 1))
print("  max :", points_allowed.max())
print("  min :", points_allowed.min())
print("  std dev :", round(points_allowed.std(), 2))
```

## Task 3 — Ask two questions with masks
Answer at least **two** real questions using boolean masks. Examples:
- Weather: How many rainy days? Which days were above average temperature?
- Sports: How many games were won? Which games scored above 100?

Print clear answers.

```python
# YOUR CODE HERE
points_scored = data["points_scored"]
# How many games won? 
points_allowed = (points_allowed > points_scored).sum()
print("Number of games won:", int(points_allowed))

```

```python
points_allowed = data['points_allowed']
points_allowed = (points_allowed > 100).sum()
print("Number of days they scored above 100:", int(points_allowed))
```

## Task 4 — Find the best/extreme row
Use `argmax` (or `argmin`) to find and print the most extreme row — e.g., the hottest day, or the highest-scoring match.

```python
# YOUR CODE HERE
# Which day the score was highest? (argmax gives the POSITION of the max)
points_allowed = data['points_allowed']
highest_pos = points_allowed.argmax()
print(highest_pos)
highest_score = data['match'][highest_pos]
print("Highest score was on day", int(highest_score), "at", points_allowed[highest_pos])
```

## Task 5 — Label with np.where
Create a label array with `np.where` (e.g., HOT/MILD, or WIN/LOSS-style labels). Print it neatly.

```python
# YOUR CODE HERE
# Label each day WIN / LOSS using np.where
labels = np.where(points_allowed > points_scored, "WIN", "LOSS")
for m, p, lab in zip(data['match'].astype(int), points_allowed, labels):
    print(f"Match {m:2d}: {p:.0f} -> {lab}")
```
